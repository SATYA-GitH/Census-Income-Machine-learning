```python
adult = pd.read_csv('adult.data.csv')

adult.head(30)
```

```python
adult['workclass'].value_counts()
```

```python
# Filling the values with Mode value which is Private( highest) for workclass

# stripping 

adult['workclass'] = adult['workclass'].str.strip()

# Replace 

adult['workclass'] = adult['workclass'].replace('?','Private')

```

```python
adult.head(30)
```

```python
adult['occupation'] = adult['occupation'].str.strip()

adult['occupation'] = adult['occupation'].replace('?','Prof-specialty')
```

```python
adult.head(30)
```

```python
adult['native-country'].value_counts()
```

```python
adult['native-country'] = adult['native-country'].str.strip()

adult['native-country'] = adult['native-country'].replace('?','United-States')
```

```python
adult.head(30)
```

## *Successfully filled the ? - missing values with mode*
