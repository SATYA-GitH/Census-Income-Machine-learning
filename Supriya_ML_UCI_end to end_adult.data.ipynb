```python
g = sns.FacetGrid(df, col='income', row='race', hue='race')
g.map(plt.scatter, 'age', 'age')
g.add_legend()
plt.show()
```

```python
# Create a FacetGrid
g = sns.FacetGrid(df, col='income', row='marital-status', hue='marital-status', margin_titles=True)

# Define a function to plot a bar chart in each facet
def bar_plot(x, y, **kwargs):
    # Count occurrences of y values
    counts = y.value_counts()
    # Create bar plot
    plt.bar(counts.index, counts.values, **kwargs)
    plt.xlabel('income')
    plt.ylabel('Count')

# Map the bar plot function to the grid
g.map(bar_plot, 'marital-status', 'income')

# Add a legend
g.add_legend()

# Display the plot
plt.show()
```

```python
# Aggregate the data
counts = df.groupby(['income', 'native-country']).size().unstack(fill_value=0)

# Plot a stacked bar chart
counts.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Income')
plt.ylabel('Count')
plt.title('Count of Native Country by Income')
plt.legend(title='Native Country')
plt.show()
```

```python

# Aggregate the data
counts = df.groupby(['income', 'race']).size().unstack(fill_value=0)

# Plot: Donut Chart for Race and Income
fig, ax = plt.subplots(figsize=(10, 6))

# Define a color map
colors = plt.get_cmap('Set1').colors

# Plotting for each income category
for i, income in enumerate(counts.index):
    data = counts.loc[income]
    wedges, texts, autotexts = ax.pie(data, labels=data.index, autopct='%1.1f%%',
                                      colors=colors, startangle=90, wedgeprops=dict(width=0.4),
                                      textprops=dict(color="black"))
    ax.set_title(f'Donut Chart of Races for Income: {income}')

# Adjust layout to make room for the legend
plt.subplots_adjust(right=0.6)

# Add a vertical legend outside the plot with black text
plt.legend(wedges, data.index, title="Race", loc="center left", bbox_to_anchor=(1.05, 0.5, 0.4, 1), fontsize=12, title_fontsize='13', frameon=False, labelspacing=1.2, handletextpad=1.5)

# Set the legend text color to black
for text in plt.gca().get_legend().get_texts():
    text.set_color('black')

plt.show()

```
